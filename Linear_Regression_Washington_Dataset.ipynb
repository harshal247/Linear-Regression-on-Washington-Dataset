```python
import pandas as pd
import numpy as np
import math
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn import preprocessing
```

```python
df = pd.read_csv('day.csv')
df = df.drop(columns=['instant', 'dteday'])
```

```python
normalized_df = (df-df.mean())/df.std()
```

```python
normalized_df.corr()
```

```python
col = df.columns
col = col.delete(13)
```

```python
col
```

```python
corr_list = []
for c in col:
    corr_list.append(df['cnt'].corr(df[c]))
print(corr_list)    
```

```python
corr_df = pd.DataFrame(corr_list, col )
```

```python
corr_df.columns = ['cnt']
```

```python
corr_df
```

```python
y = df['cnt']
x = df.drop(columns=['cnt','casual','registered'])

X_train, X_test, Y_train, Y_test = train_test_split(x, y, train_size= 0.7, test_size=0.3, random_state=0)
```

```python
print(X_train, Y_train)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
x_pred = model.predict(X_train)
```

```python
x_pred
```

```python
model_coef = pd.DataFrame(model.coef_)
model_coef.columns = ['coefficient']
```

```python
model_coef
```

```python
r2_train = r2_score(Y_train, x_pred)
mse_train = mean_squared_error(Y_train, x_pred)
rmse_train = math.sqrt(mse_train)
```

```python
print("i) RMSE metric : ",rmse_train)
print("ii) R^2 metric : ",r2_train)
```

```python
y_pred = model.predict(X_test)
```

```python
r2_test = r2_score(Y_test, y_pred)
```

```python
mse_test = mean_squared_error(Y_test, y_pred)
rmse_test = math.sqrt(mse_test)
```

```python
print("i) RMSE metric : ",rmse_test)
print("ii) R^2 metric : ",r2_test)
```
